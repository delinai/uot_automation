## Making Classes and storing data and functions in classes

```python
class Client:

    def __init__(self, name, phone, email, prod):
        self.name = name
        self.phone = phone
        self.email = email
        self.prod = prod if prod is not None else []
    
    def print_info(self):
        print(f"Name: {self.name}, Phone: {self.phone}, Email: {self.email}, Product: {self.prod}")
    
    def count_items(self):
        print(len(self.prod))
    
    @classmethod
    def input(cls):
        prod_list = input('Products (comma separated)').split(',')

        prod_list = [prod.strip() for prod in prod_list]

        return cls(
                input('Name: '),
                (input('Phone Number: ')),
                (input('Email: ')),
                prod_list,
                )
```

```python
client1 = Client.input()
```

```python
client1.count_items()
```

```python
def earliest_delivery(day_of_week, order_date):
    if day_of_week == 'Saturday':
        return order_date + datetime.timedelta(days = 6)
    elif day_of_week == 'Sunday':
        return order_date + datetime.timedelta(days = 5)
    else:
        return order_date + datetime.timedelta(days=7)
```

```python
orders['earliest_delivery'] = orders.apply(lambda x: earliest_delivery(x['day_of_week'], x['order_date']), axis=1)
```

```python
orders['is_holiday'] = orders['earliest_delivery'].apply(lambda x: x in ca_holidays)
```

```python
def final_delivery_date(is_holiday, earliest_delivery):
    if is_holiday == True:
        return earliest_delivery + datetime.timedelta(days=1)
    else:
        return earliest_delivery
```

```python
orders['delivery_date'] = orders.apply(lambda x: final_delivery_date(x['is_holiday'], x['earliest_delivery']), axis=1)
```
