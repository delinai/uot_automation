```python
# in a list we can find information in 2 ways: look by index position, or search by value
list1 = [1,2,5,6,2]

# to find the value 2 we can search by a specific index position if we know what that position is
print(list1[1])

# or use the index() function to find the index position, and put that inside the slice to return the value we're looking for
print(list1.index(2))

# the limitation is if the value occurs in more than 1 position, the function will only return the first instance
print(list1[list1.index(2)])
```

```python
# compared to lists, dataframes offer more flexibility
# the loc[] function allows us to find a specific value and returns all observations which meet the condition
cust_df.loc[]

# the iloc[] function lets us find a value based on index, similar to lists
cust_df.iloc[]
```

```python
avg_purchase = cust_df['PurchaseAmount'].mean()
```

```python
avg_purchase
```

```python
# I can add a new column which holds the Avg Purchase amount value, and compare whether customers are purchase more or less than the average

new_df = cust_df[['CustomerID','PurchaseAmount']]
```

```python
new_df['AvgPurchAmount'] = avg_purchase
```

```python
new_df['Above_or_Below'] = new_df['PurchaseAmount'] - new_df['AvgPurchAmount']
```

```python
new_df
```

```python
# sorting, aggregating
# we can sort the dataframe along the index using the sort_index(), or along a column(s) using the sort_values() function

new_df.set_index('CustomerID').sort_index()
```

```python
# with axis=1 we can use the sort_values() function to sort across columns
new_df.sort_values(9, axis=1)
```

```python
# with axis=0 we can use the sort_values() function to sort down a column
# set the ascending parameter to False to sort in reverse
# set the inplace parameter to True to make the change permanent

new_df.sort_values('CustomerID', axis=0, ascending = False, inplace=True)
```

```python
# use iloc to return values based on index position
new_df.set_index('CustomerID').iloc[0:5]
```

```python
# the alternative to filtering on an index value, is to filter a column
# write a condition and put it inside square brackets

new_df[new_df['CustomerID'] == 1002]
```

```python
# filter all purchase amounts over 50
new_df[new_df['PurchaseAmount']>50]
```

```python
# filtering a specific condition > let's say customer 1002 AND they have transactions over $70

new_df[(new_df['CustomerID'] == 1002) & (new_df['PurchaseAmount'] > 70)]
```

```python
# aggregation - let's find the total purchase amount by customer
# the basic syntax is: df.groupby('column_name_of_column_youre_grouping')['column_to_aggregate'].aggfunction()

new_df.groupby('CustomerID')['PurchaseAmount'].sum()
```

```python
#let's find the average purchase amount

new_df.groupby('CustomerID')['PurchaseAmount'].mean()
```

```python
# we can add the to_frame() function to convert a series to a DataFrame

new_df.groupby('CustomerID')['PurchaseAmount'].mean().to_frame()
```

```python
# the output will automatically be a dataframe if you aggregate more than 1 column

new_df.groupby('CustomerID')[['PurchaseAmount','Above_or_Below']].mean()

# in the output, we can see that on AVERAGE, customer 1002 purchases about $76 worth of product, and is about $3.70 BELOW the average transaction amount of all customers
```

```python
import numpy as np
# we can also aggregate multiple metrics

new_df.groupby('CustomerID')[['PurchaseAmount','Above_or_Below']].agg(['mean','sum'])

# now I can see the average of each column by customer, AND the total of each column BY customer
```

```python
# aggregate columns in different ways using a dictionary

new_df.groupby('CustomerID')[['PurchaseAmount','Above_or_Below']].agg({'PurchaseAmount':'sum','Above_or_Below':'mean'})

```
